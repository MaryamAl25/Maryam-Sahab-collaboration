```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("Student Depression Dataset.csv")
df
```

```python
df.drop(columns=['Work Pressure'], inplace=True)
df.drop(columns=['Study Satisfaction'], inplace=True)
df.drop(columns=['Job Satisfaction'], inplace=True)
df.drop(columns=['Have you ever had suicidal thoughts ?'], inplace=True)
df.drop(columns=['Family History of Mental Illness'], inplace=True)
df.drop(columns=['Financial Stress'], inplace=True)
```

```python
#Find averages, maximums, minimums of numeric columns 
```

```python
#the average performance score of each student 

df["Average_Score"] = df[["CGPA", "Academic Pressure", "Work/Study Hours"]].mean(axis=1)
df.head(10)
```

```python
# the maximum CGPA
max_CGPA = df["CGPA"].max()
print("The maximum CGPA:",max_CGPA)
```

```python
higher_CGPA_Student = df[df["CGPA"] == df["CGPA"].max()]
higher_CGPA_Student
```

```python
# the minimum CGPA
min_CGPA = df["CGPA"].min()
min_CGPA
```

```python
lowest_CGPA_Student = df[df["CGPA"] == df["CGPA"].min()]
lowest_CGPA_Student
```

```python
#Filter rows using conditions (.loc[], .isin(), .between()) 
```

```python
# using isin will return the student that are from "Delhi", "Hyderabad","Pune" only

df[df["City"].isin(["Delhi", "Hyderabad","Pune"])]
```

```python
# age of student that are between 20,25
df[df["Age"].between(20,25)]
```

```python
#5. Perform data visualization using matplotlib: 
```

```python
#Line plot: Show trend of one numeric variable over another (e.g., sales over months) 
```

```python
# line plot for Academic Pressure
plt.plot(df["CGPA"], df["Academic Pressure"], marker = 0, linestyle="-", color='red')

plt.title("Academic Pressure of students")
plt.xlabel("Student CGPA")
plt.ylabel("Academic Pressure")
plt.grid(True)
plt.show()
```

```python
# Bar chart: Compare numeric values across categories 
```

```python
#Bar  plot for Sleep Duration & Academic Pressure	
plt.bar(df["Academic Pressure"],df["Sleep Duration"] ,color="blue" , width=0.4)
plt.title("Academic Pressure vs Sleep Duration")
plt.xlabel("Students Academic Pressure")
plt.ylabel("Students Sleep Duration")
plt.show()
```

```python
#Histogram: Show distribution of a numeric variable 
```

```python
#Horizontal  plot for CGPA
plt.barh(df["City"],df["CGPA"] ,color="green")
plt.title("CGPA vs location")
plt.xlabel("Students CGPA")
plt.ylabel("Students location")
plt.show()
```

```python
#Pie chart: Show proportion of categories
```

```python
#pie chart  for  student distribution by location
df["City"].value_counts().plot.pie(autopct='1.1f%%', startangle=0)
plt.title("Student by City ")
plt.ylabel("")
plt.show()
```

```python
#Scatter plot: Show relationship between two numeric variables 
```

```python
#Scatter plate : Academic Pressure & Work/Study Hours	

plt.scatter(df["Academic Pressure"], df["Work/Study Hours"], color="blue", marker="o", alpha=0.3)
plt.title("Scatter Plot: Academic Pressure vs Work/Study Hours")
plt.xlabel("Academic Pressure")
plt.ylabel("Work/Study Hours")

plt.show()
```

```python
#Box plot: Show spread and outliers for numeric columns
```

```python
#Box plot
plt.boxplot([df["Average_Score"]], labels=["Average_Score"])
plt.title("Average_Score Distribution by Subject")
plt.ylabel("Marks")
plt.show()
```

```python
#Correlation heatmap: Show correlation between numeric columns 
```

```python
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between numeric columns ")
plt.show()
```
